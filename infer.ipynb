```python
import cv2
from ultralytics import YOLO
```

```python
model = YOLO('best.pt')

# opening cam
cap = cv2.VideoCapture(0)

# Loop through each frame from the webcam
while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(frame, show=False)  # Use show=False for manual rendering

    # Display the frame with bounding boxes
    annotated_frame = results[0].plot()  # Annotate the frame with the predictions
    cv2.imshow('YOLOv8 Real-time Detection', annotated_frame)

    # q to close the webcam
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```
